```python
from PIL import Image
import face_recognition
import os

# listing = os.listdir(path1)
c=0; total=0
path="../input/att-face-dataset/ATT Face Dataset/"
listing2=os.listdir(path)
for k in listing2:
    # Access the 1st image in the subdirectory
    inner_path = path + k + "/"
    list21 = os.listdir(inner_path)
    print("\t",list21[0])
    known_image = face_recognition.load_image_file(inner_path+list21[0]) # extract the 1st image of directory
    original_encoding = face_recognition.face_encodings(known_image)[0]
    listing3=os.listdir(inner_path)

    for i in listing3: # Loop to extract all images from the directory
        print('Using image',i)
        image = face_recognition.load_image_file(inner_path+i)
        face_locations = face_recognition.face_locations(image)
        if face_locations:
            unknown_image = face_recognition.load_image_file(inner_path+i)
            unknown_encoding = face_recognition.face_encodings(unknown_image)[0]
            results = face_recognition.compare_faces([original_encoding], unknown_encoding, tolerance=0.6)
            print(results[0])
            total+=1
            if(results[0]==True):
                c+=1
        else:
            total+=1
"""    
for i in listing:
    unknown_image = face_recognition.load_image_file(path1+i)
    unknown_encoding = face_recognition.face_encodings(unknown_image)[0]
    results = face_recognition.compare_faces([original_encoding], unknown_encoding, tolerance=0.6)
    print(results[0])
"""
print("Successfully verified",c)
print("Total Images",total)
print((c/total)*100)
```
